```python
from transformers import pipeline

translator = pipeline("translation", model="Helsinki-NLP/opus-mt-ko-en")
text="내 돈 언제 갚을거야. 어제까지 준다고 해 놓고 너 정말 그러면 나쁜놈이야"
t_text=translator(text)
print(type(t_text))
summ = pipeline("summarization")
summ_text=summ(str(t_text))
print(summ_text)
```

```python
print(summ_text[0]['summary_text'])
```

```python
clf = pipeline("sentiment-analysis")

out=clf(summ_text[0]['summary_text'])
print(out)
```
