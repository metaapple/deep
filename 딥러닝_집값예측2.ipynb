```python
import torch
import torch.nn as nn

# 1) 재현성
torch.manual_seed(42)
```

```python
# 2) 예시 데이터 (면적m^2 -> 집값)
#    실제로는 CSV에서 가져오면 되지만, "가장 간단" 버전이라 여기선 직접 만듦
x = torch.tensor([[20.0],
                  [30.0],
                  [40.0],
                  [50.0],
                  [60.0]], dtype=torch.float32)   # (N, 1)

y = torch.tensor([[200.0],
                  [300.0],
                  [410.0],
                  [520.0],
                  [610.0]], dtype=torch.float32)  # (N, 1)  (예: 단위=만원/백만원 등)
```

```python
model = nn.Linear(1, 1) # 선형회귀
```

```python
loss_fn = nn.MSELoss()
opt = torch.optim.SGD(model.parameters(), lr=0.0001)
```

```python
for epoch in range(3000):
    pred = model(x)
    loss = loss_fn(pred, y)

    opt.zero_grad()
    loss.backward()
    opt.step()
```

```python
with torch.no_grad():
  new_x = torch.tensor([[45.0]], dtype=torch.float32)
  new_pred = model(new_x)
```

```python
import torch
import torch.nn as nn
torch.manual_seed(42)
```

```python
# 1) 데이터: (N, 2)
X = torch.tensor([
    [30.0, 1.0],
    [40.0, 2.0],
    [50.0, 2.0],
    [60.0, 3.0],
    [70.0, 3.0],
    [80.0, 4.0],
], dtype=torch.float32)

# 정답: (N, 1)
y = torch.tensor([
    [260.0],
    [390.0],
    [450.0],
    [580.0],
    [650.0],
    [780.0],
], dtype=torch.float32)
```

```python
model = nn.Linear(2, 1)
```

```python
loss_fn = nn.MSELoss()
opt = torch.optim.SGD(model.parameters(), lr=0.0001)
```

```python
for epoch in range(20000):
    pred = model(X)
    loss = loss_fn(pred, y)

    opt.zero_grad()
    loss.backward()
    opt.step()
```

```python
with torch.no_grad():
    test = torch.tensor([
        [55.0, 2.0],
        [75.0, 4.0],
    ], dtype=torch.float32)
    out = model(test)
```

```python
print("W:", model.weight.data)  # (1,2)
print("b:", model.bias.data)    # (1,)
print("예측값:\n", out)
```
