```python
from transformers import pipeline

translator = pipeline("translation", model="Helsinki-NLP/opus-mt-ko-en")

text1 = "안녕하세요! 저는 구글 딥마인드에서 개발한 대규모 언어 모델인 Gemma입니다."
text2 = "저는 방대한 양의 텍스트 데이터를 학습하여 만들어졌으며, 질문에 답변하고, 텍스트를 생성하고, 다양한 종류의 작업을 수행할 수 있습니다."
text3 = "저의 모델은 1.76 trillion 개의 파라미터를 가지고 있으며, 다양한 훈련 과정을 거쳤습니다. 하지만 저는 끊임없이 발전하고 있으며, 새로운 정보를 학습하고 개선될 예정입니다.."

text = text1 + text2 + text3
print(translator(text, max_length=64))
```
